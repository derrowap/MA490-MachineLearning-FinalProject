```python
from __future__ import print_function
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
df_real = pd.read_csv('Data\\adder_pred_real.csv', header=None)
df_real = df_real.T
df_real.columns=['in', 'out']

df_100 = pd.read_csv('Data\\adder_pred_100_steps.csv', header=None)
df_100 = df_100.T
df_100.columns=['in', 'out']

df_10000 = pd.read_csv('Data\\adder_pred_10000_steps.csv', header=None)
df_10000 = df_10000.T
df_10000.columns=['in', 'out']
```

```python
plt.plot(df_100['in'], df_100['out'], label='100 steps')
plt.plot(df_10000['in'], df_10000['out'], label='1000 steps')
plt.plot(df_real['in'], df_real['out'], label='Real')
plt.legend()
plt.show()
```
